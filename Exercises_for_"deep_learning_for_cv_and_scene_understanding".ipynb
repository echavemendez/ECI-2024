You will explore different networks as instructed below to see how each of them can capture (or not) the properties of the true underlying function.

```python
# importing neccessasry libraries
import numpy as np
import torch.optim as optim
from matplotlib import pyplot as plt
import torch
import torch.nn as nn
```

```python
# creating data for training

def func(x):
  return x**3 - 0.5 * x**2 - x + 1.5

x = np.linspace(-3, 3, 80)
y = func(x)

x_test =np.concatenate([np.linspace(3.1, 4, 10), np.linspace(-2.5, 2.5, 17)])
y_test = func(x_test)
x, x_test
```

```python
#@title Visualizing the ground truth training data
plt.figure()
plt.scatter(x, y)
plt.show()
```

```python
#@title TODO: Define the neural networks

class NeuralNet_1Layer(nn.Module):
  def __init__(self):
    super(NeuralNet_1Layer, self).__init__()
    # TODO define 1 layer neural network
    self.layer_1=nn.Linear(1,1)
  # TODO implement forward function
  def forward(self, x):
    out = self.layer_1(x)
    return out

class NeuralNet_2Layer(nn.Module):
  def __init__(self):
    super(NeuralNet_2Layer, self).__init__()
    # TODO define 2 layer neural network. Use 5 neurons for the hidden layer
    self.layer_hidden=nn.Linear(1,5)
    self.layer_final=nn.Linear(5,1)

  # TODO implement forward function
  def forward(self, x):
    out = self.layer_final(self.layer_hidden(x))
    return out


class NeuralNet_2Layer_activation(nn.Module):
  def __init__(self):
    super(NeuralNet_2Layer_activation, self).__init__()
    # TODO define 2 layer neural network with an activation between the layers
    # choose between common activation functions from the lecture, e.g. sigmoid, relu
    # Also here use 5 neurons for the hidden layer
    self.layer_hidden=nn.Linear(1,5)
    self.ReLU=nn.ReLU()
    self.layer_final=nn.Linear(5,1)

  # TODO implement forward function
  def forward(self, x):
    out=self.layer_final(self.ReLU(self.layer_hidden(x)))
    return out
```

```python
# convert input and output to torch tensors, so that we can feed them to a PyTorch network
x_tensor = torch.from_numpy(x).reshape(-1, 1).float()
y_tensor = torch.from_numpy(y).reshape(-1, 1).float()
x_test_tensor = torch.from_numpy(x_test).reshape(-1, 1).float()
y_test_tensor = torch.from_numpy(y_test).reshape(-1, 1).float()

# Hyper parameters
epochs = 500
batch_size = 20
learning_rate = 0.01

# initialize networks and set them to training mode
net1 = NeuralNet_1Layer().train()
net2 = NeuralNet_2Layer().train()
net3 = NeuralNet_2Layer_activation().train()

# TODO initialize an optimizer (e.g. SGD) of each net. Don't forget to pass the learning_rate as argument!
# Hint try reducing the learning rate if the loss fuction = nan for a certain model
optimizer1 = optim.SGD(net1.parameters(), lr=0.01, momentum=0.9)
optimizer2 = optim.SGD(net2.parameters(), lr=0.001, momentum=0.9)
optimizer3 = optim.SGD(net3.parameters(), lr=0.0004, momentum=0.75)

def training_loop(net, optimizer, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor):
  # TODO implement the training loop. make sure it contains all the steps from the lecture!
  for epoch in range(epochs):
    for i in range(0,len(x_tensor),batch_size):
      # TODO reset the gradients to zero
      optimizer.zero_grad()
      # TODO get current data batch
      x_input = x_tensor[i:i+batch_size]
      y_gt =y_tensor[i:i+batch_size]
      # TODO get network prediction for input batch x_input
      y_predicted = net(x_input)

      # TODO implement a reasonable loss function of your choice. keep in mind, is this a regression or a classification task?
      loss = nn.MSELoss()(y_predicted,y_gt)
      # TODO backpropagate the loss
      loss.backward()

      # TODO call an optimization step
      optimizer.step()

    # TODO evaluate the network every 10 epochs
    if epoch % 10 == 0:
      y_predicted = net(x_tensor)
      error = nn.MSELoss()(y_predicted,y_tensor).item()
      print("Epoch=", epoch, "Error=", error)

  # setting the network to evaluate mode
  net.eval()

  # TODO compute the mean L2 error on the whole training set
  y_predicted = net(x_tensor)
  error = nn.MSELoss()(y_predicted,y_tensor)
  print("Final TRAIN error=", error)

  # TODO similarly compute the mean L2 error in the test set
  y_test_predicted = net(x_test_tensor)
  error = nn.MSELoss()(y_test_predicted,y_test_tensor)
  print("Final TEST error=", error)

  # converting the predictions to numpy so that we can visualize them
  y_pred = y_predicted.reshape(-1).detach().numpy()
  y_test_pred = y_test_predicted.reshape(-1).detach().numpy()

  # vizualize predictions and ground truth data for training and test set
  plt.figure()
  plt.scatter(x, y)
  plt.scatter(x, y_pred)
  plt.scatter(x_test, y_test)
  plt.scatter(x_test, y_test_pred)
  plt.show()

# call the training loop for each of your networks
print("Starting training for 1 layer NN")
training_loop(net1, optimizer1, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor)
print("Starting training for 2 layer NN")
training_loop(net2, optimizer2, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor)
print("Starting training for 2 layer NN with activation")
training_loop(net3, optimizer3, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor)

```
